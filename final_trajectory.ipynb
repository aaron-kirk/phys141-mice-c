```python
import matplotlib.pyplot as plt
import numpy as np

n = np.linspace(-np.pi,np.pi,100000)
R_o = .25
p = 2 * R_o
u = 2.25
def orbit(n):
    x = p * n
    y = .5 * p * (1 - n**2)
    #velocity
    r = np.sqrt(x**2 +y**2)
    v = np.sqrt(2*u/r)
    vx = v * np.cos(n)
    vy = v *np.sin(n)
    return x , y, vx, vy

#full path
x1, y1, vx1, vy1 = orbit(n)
x2, y2, vx2, vy2 = orbit(n)
#intial pos and vel
x_s1, y_s1, vx_s1, vy_s1 = orbit(np.radians(-30))
x_s2, y_s2, vx_s2, vy_s2 = orbit(np.radians(-60))

#center mass m1 / (m1+m2) or m2 / (m1+m2) --> .5 in every casefor m1=m2=1
x1 = x1/2
y1 = y1/2
x_s1 = x_s1/2
y_s1 = y_s1/2
x2 = -x2/2
y2 = -y2/2
x_s2 = -x_s2/2
y_s2 = -y_s2/2
vx1 = vx1/2
vy1 = vy1/2
vx_s1 = vx_s1/2
vy_s1 = vy_s1/2
vx_s2 = -vx_s2/2
vy_s2 = -vy_s2/2
vx2 = -vx2/2
vy2 = -vy2/2

plt.plot(x1,y1)
#plt.plot(x[200],y[200],'o')
#plt.plot(0,0,'o')
plt.plot(x2,y2)
plt.plot(x_s1,y_s1,'o',label='Gal 1')
plt.plot(x_s2, y_s2,'o',label='Gal 2')
plt.legend()
print('Gal 1 vel:',vx_s1,vy_s1,'\n Gal 2 vel:',vx_s2,vy_s2)
```
